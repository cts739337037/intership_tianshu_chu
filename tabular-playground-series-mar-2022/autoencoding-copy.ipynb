```python

from bayes_opt import BayesianOptimization
```

```python
def rf_cv(cells1, cells2, kernelsize, activation1, max_delta_step, subsample, colsample_bytree,reg_alpha,reg_lambda,scale_pos_weight,gamma):
    model = Sequential()
    model.add(Conv1D(cells1, kernel_size=kernelsize, activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same', input_shape=(None, 1)))
    model.add(Conv1D(cells2, kernel_size=kernelsize,activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same'))
    model.add(Conv1D(cells2, kernel_size=kernelsize, activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same' ))
    model.add(Conv1D(cells1, kernel_size=kernelsize,activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same'))
    model.add(Conv1D(1,kernel_size=kernelsize,activation='sigmoid', padding='same'))
    model.compile(optimizer='adam' , loss='mse')
    return model
```

```python
rf_bo = BayesianOptimization(
        rf_cv,
                {'cells1':(128,256),
                'cells2': (16,128),
                'n_estimators': (200,1500),
                'min_child_weight': (0,20),
                'max_delta_step': (0.01,2),
                'subsample':(0.6,1),
                'colsample_bytree': (0.5,1),
                'reg_alpha': (0.01,1),
                'reg_lambda': (0.2,1),
                'scale_pos_weight': (0.2,1),
                'gamma':(0,0.5)}
    )
```

```python
def set_model_encoding_mm():
    model = Sequential()
    model.add(Conv1D(128, kernel_size=3, activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same', input_shape=(None, 1)))
    model.add(Conv1D(32, kernel_size=3,activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same'))
    model.add(Conv1D(32, kernel_size=3, activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same' ))
    model.add(Conv1D(128, kernel_size=3,activation='elu', kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform',padding='same'))
    model.add(Conv1D(1,kernel_size=3,activation='sigmoid', padding='same'))
    model.compile(optimizer='adam' , loss='mse')
    
    return model
modelen_mm = set_model_encoding_mm()
modelen_mm.summary()
```

```python
callback = EarlyStopping(monitor='loss', patience=10)
history_mm = modelen_mm.fit(X_train_mm, Y_train_mm, validation_data = (X_val_mm, Y_val_mm), epochs=400, batch_size=32, verbose=1, callbacks=[callback])
```

```python
def features(dataset):
    ret_dataset=pd.DataFrame(index=dataset.index)
    for i in range(2,52,1):
        window = i
        # Volatility
        ret_dataset[str('SMA'+str(i))] = talib.SMA(dataset['congestion'],timeperiod=window).shift(1)
        ret_dataset[str('EMA'+str(i))] = talib.EMA(dataset['congestion'],timeperiod=window).shift(1)
        ret_dataset[str('STDDEV'+str(i))] = talib.STDDEV(dataset['congestion'],timeperiod=window).shift(1)
        ret_dataset[str('Lag'+str(i))] = dataset['congestion'].shift(i)
    return ret_dataset
```
